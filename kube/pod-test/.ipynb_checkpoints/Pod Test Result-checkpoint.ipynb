```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

# 実験1: Pod, Node数における通信時間の変化

## 実験結果

```python
# Node1
x1 = [4, 9, 16]
y1 = [23, 61, 150]
ymin1 = [18,51,109]
ymax1 = [32,78,194]
yerr1 = [np.array(y1)-np.array(ymin1), np.array(ymax1)-np.array(y1)]
label1 = "Node1"

# Node2
x2 = [4, 9, 16]
y2 = [15, 36, 86]
ymin2 = [14,28,56]
ymax2 = [18,55,100]
yerr2 = [np.array(y2)-np.array(ymin2), np.array(ymax2)-np.array(y2)]
label2 = "Node2"


# Node3
x3 = [4, 9, 16, 25]
y3 = [12, 33, 44, 95]
ymin3 = [10,24,36, 57]
ymax3 = [15,48,54, 125]
yerr3 = [np.array(y3)-np.array(ymin3), np.array(ymax3)-np.array(y3)]
label3 = "Node3"


# Node4
x4 = [4, 9, 16, 25]
y4 = [9, 24, 42, 56]
ymin4 = [8,19,23, 38]
ymax4 = [11,33,65,78]
yerr4 = [np.array(y4)-np.array(ymin4), np.array(ymax4)-np.array(y4)]
label4 = "Node4"

fig, ax = plt.subplots()
ax.errorbar(x1, y1,marker="o", color = "blue",  linestyle = "-", yerr=yerr1, capsize=3, label=label1)
ax.errorbar(x2, y2,marker="o", color = "red",  linestyle = "-", yerr=yerr2, capsize=3, label=label2)
ax.errorbar(x3, y3,marker="o", color = "orange",  linestyle = "-", yerr=yerr3, capsize=3, label=label3)
ax.errorbar(x4, y4,marker="o", color = "green",  linestyle = "-", yerr=yerr4, capsize=3, label=label4)
ax.legend(loc=0)    # 凡例
plt.xlabel("WorkerPod数")
plt.ylabel("1サイクルにおける通信時間(ms)")
plt.savefig("lc-test.png",format = 'png', dpi=300)
plt.show()
```

## 考察

### Pod数と通信時間の推移について
WorkerPodの増加すればするほど、通信時間の増加率があがる。指数関数的に増加していく。

### Node数と通信時間の推移について
Node数を増加させることによって、通信時間を短縮することができている。ある一定のところで通信時間の短縮がほとんどなくなる。
worker数9であればNode2で十分なリソースを確保できていて、worker数16であればNode数3で十分なリソースを確保できていると言える。
